```python
sns.countplot(x='Churn', data=df)

plt.title('Customer Churn Distribution')
plt.xlabel('Churn Status')
plt.ylabel('Number of Customers')

plt.show()
```

```python
sns.countplot(x='Contract', hue='Churn', data=df)

plt.title('Churn by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Number of Customers')

plt.xticks(rotation=45)

plt.show()
```

```python
sns.histplot(data=df, x='tenure', hue='Churn', bins=30)

plt.title('Customer Tenure Distribution by Churn')
plt.xlabel('Tenure (Months)')
plt.ylabel('Number of Customers')

plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)

plt.title('Monthly Charges vs Churn')
plt.xlabel('Churn Status')
plt.ylabel('Monthly Charges')

plt.show()
```

```python
sns.countplot(x='InternetService', hue='Churn', data=df)

plt.title('Churn by Internet Service')
plt.xlabel('Internet Service')
plt.ylabel('Number of Customers')

plt.show()
```

Business Insights

1. Customer Churn Rate:
- The overall churn rate is approximately 26.58%, meaning around 1 in 4 customers leave the company.
- Reducing churn can help improve customer retention and revenue.

2. Contract Impact on Churn:
- Month-to-month contract customers have the highest churn rate.
- Customers with one-year and two-year contracts show better retention.
- Offering incentives for longer-term contracts may reduce churn.

3. Customer Tenure Analysis:
- Customers with shorter tenure are more likely to churn.
- Improving onboarding and early customer experience can help retain new customers.

4. Monthly Charges Impact:
- Customers with higher monthly charges show a higher tendency to churn.
- Reviewing pricing plans and offering better value can help reduce customer exits.

5. Internet Service Analysis:
- Churn varies across different internet service categories.
- The company should investigate service quality and customer satisfaction for high-churn segments.

6. Recommendations:
- Focus on retaining new customers during their first few months.
- Encourage customers to move from month-to-month plans to long-term contracts.
- Provide personalized offers to high-risk customers.
- Improve services for segments with higher churn rates.
